```python
#3 comb
def read_combine_three(p_numtr,p_numte, p_ctgtr, p_ctgte, p_dtr, p_dte, ctg_col,  num_col, d_col):
    
    X_par_train_ctg = pd.read_csv(p_ctgtr, usecols=ctg_col)
    X_par_test_ctg  = pd.read_csv(p_ctgte, usecols=ctg_col)
    
    #encode all and split (need the same standard)
    encoded=pd.get_dummies(pd.concat([X_par_train_ctg,X_par_test_ctg]))
    trainrows=X_par_train_ctg.shape[0]
    X_train_ctg=encoded.iloc[:trainrows, : ]
    X_test_ctg=encoded.iloc[trainrows: , : ]  
    
    print('####################### Categorical Data Ready #####################')
    
    #numeric 
    X_train_num=pd.read_csv(p_numtr, usecols=num_col)
    X_test_num=pd.read_csv(p_numte, usecols=num_col)
    
    print('####################### Numeric Data Ready ########################')
    
    X_train_d=pd.read_csv(p_dtr, usecols=d_col)
    X_test_d=pd.read_csv(p_dte, usecols=d_col)
    
    print('########################  Date Data Ready  ########################')
    
    response=pd.read_csv(p_numtr,usecols=[969])
    
    combined_inter_tr=pd.merge(X_train_ctg, X_train_num, on='Id')
    
    combined_train=pd.merge(X_train_d, combined_inter_tr, on='Id')
    
    combined_inter_te=pd.merge(X_test_ctg, X_test_num, on='Id')
    
    combined_test=pd.merge(X_test_d, combined_inter_te, on='Id')
    
    print('####################### Combined Data Ready ########################')
    
    return combined_train, response, combined_test
```

```python
def mcc_scorer(predd,yd):
    #pred=predd.get_label()
    pred=predd.tolist()
    print("min: %.15f  max:%.15f"%(min(pred),max(pred)))
    y=yd.get_label()
    
    
    n=len(pred)
    max_v=(min(pred)+0.5)/2
    min_v=max([min(pred)+0.00001, 0.15])
    can=list(arange(min_v,max_v,0.002))
    
    len_can=len(can)
    
    # number of positive values
    num_p=int(sum(y))
    # number of negtive values
    num_n=len(y)-num_p
    
    tp=[sum([1 for i in range(n) if pred[i]>=c and y[i]==1]) for c in can]
    tn=[sum([1 for i in range(n) if pred[i]<c and  y[i]==0]) for c in can]
    
    #fp=[sum([1 for i in range(n) if pred[i]>=c and y[i]==0]) for c in can]
    #fn=[sum([1 for i in range(n) if pred[i]<c and  y[i]==1]) for c in can]
    fn=[num_p-ele for ele in tp]
    fp=[num_n-ele for ele in tn]
    

    score=[(tp[i]*tn[i]-fp[i]*fn[i])/(((tp[i]+fp[i])*(tp[i]+fn[i])*(tn[i]+fp[i])*(tn[i]+fn[i]))**(1/2.0))
           for i in range(len_can)]
    fmax=max(score)
    p=score.index(fmax)
    threshold=can[p]
    print('threshold:',threshold)
    print('TP:%d,TN:%d,FP:%d,FN:%d'%(tp[p],tn[p],fp[p],fn[p]))
    return('mcc',fmax)


```

```python
def xgb_train_watch(combined_train, response):
    X_dtrain, X_deval, y_dtrain, y_deval = train_test_split(combined_train, response,
                                                                             random_state=2046, test_size=0.3)
    dtrain = xgb.DMatrix(X_dtrain, y_dtrain)
    deval = xgb.DMatrix(X_deval, y_deval)
    watchlist = [(deval, 'eval')]
    params = {
        'booster': 'gbtree',
        'objective': 'binary:logistic',
        'colsample_bytree': 0.8,
        'eta': 0.1,
        'max_depth': 10,
        'seed': 2017,
        'silent': 0,
       # 'gamma':0.005,
        'subsample':0.8
    }
    
    clf = xgb.train(params, dtrain, 100, watchlist, early_stopping_rounds=50,feval=mcc_scorer,maximize=True)
    
    return clf
```

```python
def gridcv(combined_train,response):
    mcc_scorer = make_scorer(matthews_corrcoef)

    grid = {'max_depth':[4,9,10,11],
        'subsample': [0.8]
 #       'colsample_bytree':[0.8,0.9],
           }
    params = {
       # 'booster': 'gbtree',
        'objective': 'binary:logistic',
        'colsample_bytree': 0.8,
        'learning_rate': 0.1,
        'seed': 2016,
        'n_estimators':100,
        'silent': 0
     }

    clf=xgb.XGBClassifier()
    clf.set_params(**params)
    #i.e. model
    search=GridSearchCV(estimator=clf,param_grid=grid, scoring=mcc_scorer,n_jobs=1,refit=True,cv=10,verbose=100)
    
    search.fit(combined_train,response['Response'])
    print(search.best_params_)
    print(search.best_score_)
    
    return search
#search.predict()
```
