```python
#not so efficient one
def my_mcc_scorer(predd,yd):
    #pred=predd.get_label()
    pred=predd.tolist()
    print("min: %.15f  max:%.15f"%(min(pred),max(pred)))
    y=yd.get_label()
    
    
    n=len(pred)
    max_v=(min(pred)+0.5)/2
    min_v=max([min(pred)+0.00001, 0.15])
    can=list(arange(min_v,max_v,0.002))
    
    len_can=len(can)
    
    # number of positive values
    num_p=int(sum(y))
    # number of negtive values
    num_n=len(y)-num_p
    
    tp=[sum([1 for i in range(n) if pred[i]>=c and y[i]==1]) for c in can]
    tn=[sum([1 for i in range(n) if pred[i]<c and  y[i]==0]) for c in can]
    
    #fp=[sum([1 for i in range(n) if pred[i]>=c and y[i]==0]) for c in can]
    #fn=[sum([1 for i in range(n) if pred[i]<c and  y[i]==1]) for c in can]
    fn=[num_p-ele for ele in tp]
    fp=[num_n-ele for ele in tn]
    

    score=[(tp[i]*tn[i]-fp[i]*fn[i])/(((tp[i]+fp[i])*(tp[i]+fn[i])*(tn[i]+fp[i])*(tn[i]+fn[i]))**(1/2.0))
           for i in range(len_can)]
    fmax=max(score)
    p=score.index(fmax)
    threshold=can[p]
    print('threshold:',threshold)
    print('TP:%d,TN:%d,FP:%d,FN:%d'%(tp[p],tn[p],fp[p],fn[p]))
    return('mcc',fmax)


```

```python
def mcc_calculator(tp, tn, fp, fn):
    sup = tp * tn - fp * fn
    inf = (tp + fp) * (tp + fn) * (tn + fp) * (tn + fn)
    if inf == 0:
        return 0
    else:
        return sup / np.sqrt(inf)

def mcc_scorer(y_prob, yd):
    y_true=yd.get_label()
    idx = np.argsort(y_prob)
    y_true_sort = y_true[idx]
    n = y_true.shape[0]
    nump = 1.0 * np.sum(y_true)  # number of positive
    numn = n - nump  # number of negative
    
    tp = nump
    tn = 0.0
    fp = numn
    fn = 0.0
    
    best_mcc = 0.0
    best_id = -1
    mccs = np.zeros(n)
    for i in range(n):
        if y_true_sort[i] == 1:
            tp -= 1.0
            fn += 1.0
        else:
            fp -= 1.0
            tn += 1.0
        new_mcc = mcc_calculator(tp, tn, fp, fn)
        mccs[i] = new_mcc
        if new_mcc >= best_mcc:
            best_mcc = new_mcc
            best_id = i
            ftp=tp
            ffn=fn
            ffp=fp
            ftn=tn
        else:
            continue
            
    best_proba = y_prob[idx[best_id]]
    print("Threshold", best_proba)
    
    print('TP:%d,TN:%d,FP:%d,FN:%d'%(ftp,ftn,ffp,ffn))

    return("MCC", best_mcc)
```

```python
def xgb_train_watch(combined_train, response):
    X_dtrain, X_deval, y_dtrain, y_deval = train_test_split(combined_train, response,
                                                                             random_state=2046, test_size=0.2)
    prior = 0.0058112
    dtrain = xgb.DMatrix(X_dtrain, y_dtrain)
    deval = xgb.DMatrix(X_deval, y_deval)
    watchlist = [(deval, 'eval')]
    params = {
        'booster': 'gbtree',
        'objective': 'binary:logistic',
        'colsample_bytree': 1,
        'eta': 0.135,
        'max_depth': 11,
        'seed': 2017,
        'silent': 0,
       # 'gamma':0.005,
        'base_score': prior,
        'subsample':0.8

    }
    
    print(params)
    
    clf = xgb.train(params, dtrain, 280, watchlist, early_stopping_rounds=50,feval=mcc_scorer,maximize=True)
    
    
    return clf
```

```python
def gridcv(combined_train,response):
    mcc_scorer = make_scorer(matthews_corrcoef)

    grid = {#'max_depth':[4,9,10,11],
      #  'subsample': [0.8, 0.75],
       # 'gamma':[0, 0.005]
        'learning_rate':[0.1,0.13,0.15],
        'n_estimators': [150,200,250]
       # 'colsample_bytree': [0.8, 0.75]
 #       'colsample_bytree':[0.8,0.9],
           }
    params = {
       # 'booster': 'gbtree',
        'max_depth':10,
        'objective': 'binary:logistic',
        'colsample_bytree': 0.8,
       # 'learning_rate': 0.1,
        'seed': 2016,
        #'n_estimators':100,
        'silent': 0
     }

    clf=xgb.XGBClassifier()
    clf.set_params(**params)
    #i.e. model
    search=GridSearchCV(estimator=clf,param_grid=grid, scoring=mcc_scorer,n_jobs=1, refit=True,cv=5, verbose=100)
    
    search.fit(combined_train,response['Response'])

    print("best param:",search.best_params_)
    
    means = search.cv_results_['mean_test_score']
    stds  = search.cv_results_['std_test_score']
    #print(search.best_score_)
    for mean, std, params in zip(means, stds, clf.cv_results_['params']):
        print("%0.5f (+/-%0.03f) for %r" % (mean, std * 2, params))
        
    return search
#search.predict()
```

```python
def xgbcv(combined_train, response):
    mcc_scorer = make_scorer(matthews_corrcoef)

    prior = 0.0058112

    xgb_params = {
        'seed': 0,
        'colsample_bytree': 0.8,
        'silent': 0,
        'subsample': 0.8,
        'learning_rate': 0.1,
        'objective': 'binary:logistic',
        'max_depth': 10,
       # 'num_parallel_tree': 1,
        #'min_child_weight': 2,
       # 'eval_metric': 'auc',
        'base_score': prior
    }


    dtrain = xgb.DMatrix(combined_train,label=response)
    res = xgb.cv(xgb_params, dtrain, num_boost_round=150, nfold=10, seed=0, stratified=True,
                 early_stopping_rounds=50, verbose_eval=100, show_stdv=True, feval=mcc_scorer,maximize=True)
```
